```python
def detect_main(sub) :
    if sub in 상의 :
        main = '상의'

    elif sub in 바지 :
        main = '바지'

    elif sub in 아우터 :
        main = '아우터'

    elif sub in 원피스 :
        main = '원피스'

    elif sub in 스커트 :
        main = '스커트'

    return main 
```

```python
# csv를 다 합치기

for idx, csv_file in enumerate(os.listdir('./site_crawling_csv_save')) :
    if idx == 0 : 
        total_df = pd.read_csv('./site_crawling_csv_save/' + csv_file)

    df = pd.read_csv('./site_crawling_csv_save/' + csv_file)

    total_df = pd.concat([total_df, df])


# csv에서 image_url이 겹치는 것은 중복을 제거하기
# 8424 -> 중복 제거 -> 1267

unique_image_df = total_df.drop_duplicates(subset=['image_url'])

selected_df = unique_image_df[['image_url','keyword' ]]

# 한 폴더안에 이미지를 모으기 단 저장할 때 -> ably_상의_후드_image1.jpg 이런식으로 저장될 수 있게 할 것



for idx, url in enumerate(selected_df['image_url']) :
    try :
        # 이미지를 일단 객체로서 불러오기 -> 객체 생성
        response = requests.get(url)
        response.raise_for_status()  
        img = Img.open(BytesIO(response.content))

        # 원본 사진을 그래도 저장하기
        sub = selected_df['keyword'].iloc[idx]
        main = detect_main(sub)
        img.save(f'./all_datasets/composition/ably_{main}_{sub}_image_{idx}.jpg')



    except Exception as e :
        print(f'{url}에서 이미지를 불러오지 못했습니다.')
        print(f'에러 내용 : {e}')
        print('------------------------------------------')


```
